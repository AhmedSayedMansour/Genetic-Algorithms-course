```python
import numpy as np
import os
from scipy import optimize
from scipy.io import loadmat
```

```python
file = open("thetas.txt" , "r")

# Theta 1
theta1_size = file.readline().split()
t1r = int(theta1_size[0])
t1c = int(theta1_size[1])

Theta1 = np.zeros((t1r,t1c))

theta1_values = file.readline().split()

indx = 0
for i in range(t1r):
    for j in range(t1c):
        Theta1[i][j] = float(theta1_values[indx])
        indx += 1

# Theta 2
theta2_size = file.readline().split()
t2r = int(theta2_size[0])
t2c = int(theta2_size[1])

Theta2 = np.zeros((t2r,t2c))

theta2_values = file.readline().split()

indx = 0
for i in range(t2r):
    for j in range(t2c):
        Theta2[i][j] = float(theta2_values[indx])
        indx += 1
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def predict(Theta1, Theta2, X):
   
    m = X.shape[0]
    ON = Theta2.shape[0]
    p = np.zeros(X.shape[0])
    
    hiddenLayer = np.array(sigmoid(np.dot(X,Theta1.T)))
    outputLayer = np.array(sigmoid(np.dot(hiddenLayer,Theta2.T)))

    p = outputLayer

    return p
```

```python
def Normalization(X):

    X_norm = X.copy()
    minimum = np.zeros(X.shape[1])
    valuesRange = np.zeros(X.shape[1])
    m,n = X.shape

    for i in range(n):
        #mu[i] = np.mean(X_norm[: , i])
        #sigma[i] = np.std(X_norm[: , i])
        minimum[i] = np.min(X_norm[: , i])
        valuesRange[i] = np.max(X_norm[: , i]) - np.min(X_norm[: , i])
    for i in range(m):
        for j in range(n):
            X_norm[i,j] = (X[i,j] - minimum[j])/valuesRange[j]
    
    return X_norm
```

```python
# Read From File
file = open("input.txt" , "r")

# Line 1
line1 = file.readline().split()
IN = int(line1[0])
HN = int(line1[1])
ON = int(line1[2])
                   
# Line 2
line2 = file.readline()
m = int(line2)

X = np.zeros((m, IN))
y = np.zeros((m, ON))

# Remaining Lines (X , y)
lines = file.readlines()
idx = 0
for line in lines:
    line_splited = line.split()
    for i in range(IN):
        X[idx][i] = float(line_splited[i])
    for j in range(ON):
        y[idx][j] = float(line_splited[j+IN])
    idx += 1
    
file.close()

# Predict Index
X = Normalization(X)
X = np.concatenate([np.ones((m, 1)), X], axis=1)
y = Normalization(y)

p = predict(Theta1, Theta2, X)

for i in range (len(y)):
    print("Y : " , y[i] , " P : " , p[i])
```
